```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("HumanResources.csv")
df.head(10)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
print("Total Rows:", len(df))
print("Null IDs:", df['id'].isnull().sum())
print("Unique IDs (non-null):", df['id'].nunique())
```

```python
df = df.dropna(subset=['id'])
```

```python
df.isnull().sum()
```

```python
# Find duplicates across entire row
duplicates_all = df[df.duplicated(keep=False)]

# Count duplicates
dup_count_all = df.duplicated().sum()
print(f"Number of completely duplicate rows: {dup_count_all}")
```

```python
df.info()
```

```python
df['hire_date'] = pd.to_datetime(df['hire_date'], errors='coerce', utc=True)
unique_years = df['hire_date'].dt.year.dropna().unique()
print(sorted(unique_years))
year_counts = df['hire_date'].dt.year.value_counts().sort_index()
print(year_counts)
```

```python
df['termdate'] = pd.to_datetime(df['termdate'], errors='coerce', utc=True)
unique_years = df['termdate'].dt.year.dropna().unique()
print(sorted(unique_years))
year_counts = df['termdate'].dt.year.value_counts().sort_index()
print(year_counts)

```

```python
df.to_csv("HumanResourcesCleanDataset.csv", index=False)
```
